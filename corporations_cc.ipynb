```python
import numpy as np
import pandas as pd
```

```python
corps_cc_df = pd.read_csv('Corporations/Corporations Responses/Climate Change/2018_Full_Climate_Change_Dataset.csv')
na_hq_df = pd.read_csv('Supplementary Data/Locations of Corporations/NA_HQ_public_data.csv')
corps_cc_info = pd.read_csv('Corporations/Corporations Disclosing/Climate Change/2018_Corporates_Disclosing_to_CDP_Climate_Change.csv')
```

```python
# variables to be defined
survey_year = 2018
survey_type = 'Climate Change'
```

```python
corps_cc_df.head()
```

```python
# adding headquarter location data
hq_data = na_hq_df[(na_hq_df['survey_year'] == survey_year) & (na_hq_df['theme'] == survey_type)]
hq_data = hq_data[['account_number','hq_country','address_city','address_state','address_lat','adress_long']]
hq_data = hq_data.rename(columns={'address_city':'hq_city','address_state':'hq_state','address_lat':'hq_lat','adress_long':'hq_long'})

corps_cc_cities = pd.merge(corps_cc_df, hq_data, how='left', on='account_number')
```

```python
corps_cc_cities.organization.unique()
```

```python
# adding industry data
corps_cc_info = corps_cc_info[['account_number','activities','sectors','industries','primary_activity','primary_sector','primary_industry','primary_questionnaire_sector']]

corps_cc_full = pd.merge(corps_cc_cities, corps_cc_info, how='left', on='account_number')
```

### Risks and Opportunities

```python
#(C2.3a) Provide details of risks identified with the potential to have a substantive financial or strategic impact on your business.

corps_cc_df_2_3a = corps_cc_df[corps_cc_df['question_number'] == 'C2.3a']
corps_cc_df_2_3a = corps_cc_df_2_3a[corps_cc_df_2_3a['response_value'].notnull()]

corps_cc_df_2_3a = corps_cc_df_2_3a.pivot_table(index=['account_number', 'organization', 'row_number'],
                                     columns='column_name', 
                                     values='response_value',
                                     aggfunc=lambda x: ' '.join(x)).reset_index()
corps_cc_df_2_3a
```

```python
#(C2.4) Have you identified any climate-related opportunities with the potential to have a substantive financial or strategic impact on your business?

corps_cc_df_2_4a = corps_cc_df[corps_cc_df['question_number'] == 'C2.4a']
corps_cc_df_2_4a = corps_cc_df_2_4a[corps_cc_df_2_4a['response_value'].notnull()]

corps_cc_df_2_4a = corps_cc_df_2_4a.pivot_table(index=['account_number', 'organization', 'row_number'],
                                     columns='column_name', 
                                     values='response_value',
                                     aggfunc=lambda x: ' '.join(x)).reset_index()
corps_cc_df_2_4a
```

### Targets and Performance

```python
#(C4.1) Did you have an emissions target that was active in the reporting year?

corps_cc_df[(corps_cc_df['question_number'] == 'C4.1')][['organization','response_value']].groupby('response_value').count().sort_values('organization',ascending=False)
```

<span style="color:red">Absolute target is like % reduction from past year, intensity target is like per unit of revenue</span>

```python
#(C4.1a) Provide details of your absolute emissions target(s) and progress made against those targets.

corps_cc_df_4_1a = corps_cc_df[corps_cc_df['question_number'] == 'C4.1a']
corps_cc_df_4_1a = corps_cc_df_4_1a[corps_cc_df_4_1a['response_value'].notnull()]

corps_cc_df_4_1a = corps_cc_df_4_1a.pivot_table(index=['account_number', 'organization', 'row_number'],
                                     columns='column_name', 
                                     values='response_value',
                                     aggfunc=lambda x: ' '.join(x)).reset_index()
corps_cc_df_4_1a
```

<span style="color:red">Interesting to see their % emission reductions compared to last year</span>

```python
#(C4.1b) Provide details of your emissions intensity target(s) and progress made against those target(s).

corps_cc_df_4_1b = corps_cc_df[corps_cc_df['question_number'] == 'C4.1b']
corps_cc_df_4_1b = corps_cc_df_4_1b[corps_cc_df_4_1b['response_value'].notnull()]

corps_cc_df_4_1b = corps_cc_df_4_1b.pivot_table(index=['account_number', 'organization', 'row_number'],
                                     columns='column_name', 
                                     values='response_value',
                                     aggfunc=lambda x: ' '.join(x)).reset_index()
corps_cc_df_4_1b
```
